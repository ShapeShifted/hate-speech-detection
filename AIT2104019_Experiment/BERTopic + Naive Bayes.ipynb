Naive-Bayes Model with Manual Mapping of Representation into Integers

```python
import pandas as pd
import numpy

df = pd.read_csv('/content/drive/MyDrive/Thesis/posts1352_standard_representations.csv')
```

```python
bertopic_words = []
for x in df['Topic']:
    bertopic_words.append(x)
```

```python
# obtain token list that's actually list instead of string
from ast import literal_eval

temp = []

for x in bertopic_words:
    temp.append((literal_eval(x)))

bertopic_words = temp
```

```python
# Flatten the list to find unique strings
flattened = [item for sublist in bertopic_words for item in sublist]

# Create a mapping of each unique string to a unique integer
unique_strings = list(set(flattened))
string_to_int = {string: i for i, string in enumerate(unique_strings)}

# Transform the original list using the mapping
converted_list = [[string_to_int[item] for item in sublist] for sublist in bertopic_words]

# Normalize the converted list
min_val = min(min(sublist) for sublist in converted_list)
max_val = max(max(sublist) for sublist in converted_list)

normalized_list = [[(item - min_val) / (max_val - min_val) for item in sublist] for sublist in converted_list]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(normalized_list, df['HateSpeech'], test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
y_pred = gnb.predict(x_test)
```

```python
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
#TOPIC
```

```python
from sklearn.naive_bayes import MultinomialNB
gnb2 = MultinomialNB(fit_prior=False, alpha = 2)
gnb2.fit(x_train, y_train)
```

```python
y_pred = gnb2.predict(x_test)
```

```python
from sklearn import metrics
print("Multinomial Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
#TOPIC
```

```python
from sklearn.naive_bayes import BernoulliNB
gnb3 = BernoulliNB(fit_prior=False,alpha = 0)
gnb3.fit(x_train, y_train)
```

```python
y_pred = gnb3.predict(x_test)
```

```python
from sklearn import metrics
print("Bernoulli Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
#TOPIC
```

Naive Bayes using a Tensorflow tokenizer to convert BERTopic representative documents into integers

```python
#obtain token list that's actually list instead of string
from ast import literal_eval

document_list = []
for x in df['Representative_Docs']:
    document_list.append(x)
```

```python
temp = []

for x in document_list:
    temp.append((literal_eval(x)))

document_list = temp
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding, Lambda
from tensorflow.keras.preprocessing.text import Tokenizer
import numpy as np
import matplotlib.pyplot as plt

# Convert the corpus to a sequence of integers
tokenizer = Tokenizer()
tokenizer.fit_on_texts(document_list)
x_train_sequences = tokenizer.texts_to_sequences(document_list)
```

```python
# Normalize the converted list
min_val = min(min(sublist) for sublist in x_train_sequences)
max_val = max(max(sublist) for sublist in x_train_sequences)

normalized_sequences = [[(item - min_val) / (max_val - min_val) for item in sublist] for sublist in x_train_sequences]
```

```python
x_train2, x_test2, y_train2, y_test2 = train_test_split(normalized_sequences, df['HateSpeech'], test_size=0.3)
```

```python
def is_uniform_2d_list(lst):
    if not lst:  # Check if the list is empty
        return True

    first_length = len(lst[0])  # Length of the first sublist
    for sublist in lst:
        if len(sublist) != first_length:
            return False
    return True

is_uniform_2d_list(x_train2)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train2, y_train2)
```

```python
y_pred2 = gnb.predict(x_test2)
```

```python
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test2, y_pred2)*100)
#REPRESENTATIVE DOCUMENT
```

```python
from sklearn.naive_bayes import MultinomialNB
gnb2 = MultinomialNB()
gnb2.fit(x_train2, y_train2)
y_pred2 = gnb2.predict(x_test2)
print("Multinomial Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test2, y_pred2)*100)
#REPRESENTATIVE DOCUMENT
```

```python
from sklearn.naive_bayes import BernoulliNB
gnb3 = BernoulliNB()
gnb3.fit(x_train2, y_train2)
y_pred2 = gnb3.predict(x_test2)
print("Bernoulli Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test2, y_pred2)*100)
# REPRESENTATIVE DOCUMENT
```
